# Are H3K27me3 silencers linked to genes with a lower expression

# HCC1954

```python
import pandas as pd
import pybedtools as pybed


#awk '{print $0}' /index/hg19/Homo_sapiens.GRCh37.87.chr.gtf | grep "protein_coding" | awk '{if ($3 == "gene") print "chr"$1"\t"$4"\t"$5"\t"$7"\t"$14 }' > /home/darragh/Annotation_reg_elements/hg19_protein_coding_minimal.gtf
gtf = "/home/darragh/Annotation_reg_elements/hg19_minimal.gtf"
def Identify_coding_promoters(gtf):
    hg19_gtf = pd.read_table(gtf)
    hg19_gtf.columns = ["chr","start","end", "strand", "gene"]
    #hg19_gtf = hg19_gtf[["chr","start","end", "gene"]]
    seqnames_we_like = ["chr1","chr2","chr3","chr4","chr5","chr6","chr7","chr8","chr9","chr10","chr11","chr12","chr13","chr14","chr15","chr16","chr17","chr18","chr19","chr20","chr21","chr22","chrX", "chrY"]
    hg19_gtf = hg19_gtf[hg19_gtf['chr'].isin(seqnames_we_like)]

    hg19_gtf.loc[hg19_gtf['strand'] == "+", ['start']] = hg19_gtf['start']
    hg19_gtf.loc[hg19_gtf['strand'] == "+", ['end']] = hg19_gtf['start']+1
    hg19_gtf.loc[hg19_gtf['strand'] == "-", ['start']] = hg19_gtf['end']
    hg19_gtf.loc[hg19_gtf['strand'] == "-", ['end']] = hg19_gtf['start']+1
    hg19_gtf = hg19_gtf[["chr","start","end", "gene"]]

    hg19_gtf['gene'] = hg19_gtf['gene'].str.rstrip(';')


    return(hg19_gtf)

Coding_promoter_df = Identify_coding_promoters(gtf)
Coding_promoter = pybed.BedTool.from_dataframe(Coding_promoter_df)
Coding_promoter = Coding_promoter.sort()


HCC1954_PcHiC = pd.read_table("/home/darragh/New/PcHiC/HCC1954/round_3/Arima/chicago/data/HCC1954.ibed", header=0)

#Focusing just on cis-cis interactions
HCC1954_PcHiC = HCC1954_PcHiC.loc[HCC1954_PcHiC['bait_chr'] == HCC1954_PcHiC['otherEnd_chr'] ]

HCC1954_PcHiC = pybed.BedTool.from_dataframe(HCC1954_PcHiC[['bait_chr',	'bait_start',	'bait_end', 'otherEnd_chr',	'otherEnd_start',	'otherEnd_end']])
HCC1954_PcHiC_genes = HCC1954_PcHiC.intersect(Coding_promoter, wa=True, wb=True)
HCC1954_PcHiC_genes = pd.read_table(HCC1954_PcHiC_genes.fn, header=None)
#This is the other end loop and its connected gene
HCC1954_PcHiC_genes = HCC1954_PcHiC_genes.iloc[:,[3,4,5,9]]
HCC1954_PcHiC_genes.columns = ['chr', 'start', 'end', 'gene']
HCC1954_PcHiC_genes = pybed.BedTool.from_dataframe(HCC1954_PcHiC_genes)

HCC1954_regs = pd.read_table("/home/darragh/New/Regulatory_elements/HCC1954_reg_elements.txt", sep = "\t", header=None)
HCC1954_regs.columns = ['chr', 'start', 'end', 'annotation']

HCC1954_regs_AE = pybed.BedTool.from_dataframe(HCC1954_regs.loc[HCC1954_regs['annotation'] == "HCC1954_active_enhancer"])
HCC1954_regs_S = pybed.BedTool.from_dataframe(HCC1954_regs.loc[HCC1954_regs['annotation'] == "HCC1954_H3K27me3_silencer"])

HCC1954_regs_AE_genes = HCC1954_regs_AE.intersect(HCC1954_PcHiC_genes, wa = True, wb= True)
HCC1954_regs_AE_genes = pd.read_table(HCC1954_regs_AE_genes.fn, header=None)
HCC1954_regs_AE_genes.columns = ['chr', 'start', 'end', 'anno', 'n1', 'n2','n3','gene']
HCC1954_regs_AE_genes['id'] = HCC1954_regs_AE_genes['chr'] + '_' + HCC1954_regs_AE_genes['start'].astype(str) + '_' + HCC1954_regs_AE_genes['end'].astype(str)

HCC1954_regs_S_genes = HCC1954_regs_S.intersect(HCC1954_PcHiC_genes, wa = True, wb= True)
HCC1954_regs_S_genes = pd.read_table(HCC1954_regs_S_genes.fn, header=None)
HCC1954_regs_S_genes.columns = ['chr', 'start', 'end', 'anno', 'n1', 'n2','n3','gene']
HCC1954_regs_S_genes['id'] = HCC1954_regs_S_genes['chr'] + '_' + HCC1954_regs_S_genes['start'].astype(str) + '_' + HCC1954_regs_S_genes['end'].astype(str)

HCC1954_regs_AE_genes = HCC1954_regs_AE_genes[['id', 'gene', 'anno']]
HCC1954_regs_AE_genes = HCC1954_regs_AE_genes.groupby(['anno', 'gene'])['id'].nunique().reset_index(name='count')

HCC1954_regs_S_genes = HCC1954_regs_S_genes[['id', 'gene', 'anno']]
HCC1954_regs_S_genes = HCC1954_regs_S_genes.groupby(['anno', 'gene'])['id'].nunique().reset_index(name='count')

HCC1954_distal_regs = pd.concat([HCC1954_regs_AE_genes, HCC1954_regs_S_genes])

all_RNA = pd.read_table("/home/darragh/New/CCLE_data/OmicsExpressionProteinCodingGenesTPMLogp1.csv", sep = ",", header=0)
all_RNA = all_RNA.rename(columns={'Unnamed: 0': 'id'})

HCC1954_RNA = all_RNA.loc[(all_RNA['id'] == 'ACH-000859') ]
HCC1954_RNA = HCC1954_RNA.T
HCC1954_RNA.columns = HCC1954_RNA.iloc[0]
HCC1954_RNA = HCC1954_RNA.iloc[1:]
HCC1954_RNA = HCC1954_RNA.reset_index()
HCC1954_RNA.columns = ['gene', 'expression']
HCC1954_RNA['gene'] = HCC1954_RNA['gene'].str.split().str[0]
HCC1954_RNA['z_score'] = (HCC1954_RNA['expression'] - HCC1954_RNA['expression'].mean() ) / HCC1954_RNA['expression'].std()


HCC1954_distal_regs_expression = HCC1954_distal_regs.merge(HCC1954_RNA, on="gene")
HCC1954_distal_regs_expression = HCC1954_distal_regs_expression.drop_duplicates()


#HCC1954_distal_regs_expression = HCC1954_distal_regs_expression.loc[HCC1954_distal_regs_expression['expression'] > 0]

print(HCC1954_distal_regs_expression.loc[HCC1954_distal_regs_expression['anno'] == "HCC1954_H3K27me3_silencer"]['expression'].median())
print(HCC1954_distal_regs_expression.loc[HCC1954_distal_regs_expression['anno'] == "HCC1954_active_enhancer"]['expression'].median())

# Apply pd.qcut within each group, handling duplicates
HCC1954_distal_regs_expression['quartile'] = HCC1954_distal_regs_expression.groupby('anno')['count'].transform(
    lambda x: pd.qcut(x.rank(method='first'), q=3, labels=[1, 2, 3])
)

HCC1954_distal_regs_expression['anno_quartile'] = HCC1954_distal_regs_expression['anno'] + '_' + HCC1954_distal_regs_expression['quartile'].astype(str)

display(HCC1954_distal_regs_expression)

HCC1954_distal_regs_expression.boxplot(column='z_score', by='anno_quartile', grid=False)
```

```python
import pandas as pd
import pybedtools as pybed


#awk '{print $0}' /index/hg19/Homo_sapiens.GRCh37.87.chr.gtf | grep "protein_coding" | awk '{if ($3 == "gene") print "chr"$1"\t"$4"\t"$5"\t"$7"\t"$14 }' > /home/darragh/Annotation_reg_elements/hg19_protein_coding_minimal.gtf
gtf = "/home/darragh/Annotation_reg_elements/hg19_minimal.gtf"
def Identify_coding_promoters(gtf):
    hg19_gtf = pd.read_table(gtf)
    hg19_gtf.columns = ["chr","start","end", "strand", "gene"]
    #hg19_gtf = hg19_gtf[["chr","start","end", "gene"]]
    seqnames_we_like = ["chr1","chr2","chr3","chr4","chr5","chr6","chr7","chr8","chr9","chr10","chr11","chr12","chr13","chr14","chr15","chr16","chr17","chr18","chr19","chr20","chr21","chr22","chrX", "chrY"]
    hg19_gtf = hg19_gtf[hg19_gtf['chr'].isin(seqnames_we_like)]

    hg19_gtf.loc[hg19_gtf['strand'] == "+", ['start']] = hg19_gtf['start']
    hg19_gtf.loc[hg19_gtf['strand'] == "+", ['end']] = hg19_gtf['start']+1
    hg19_gtf.loc[hg19_gtf['strand'] == "-", ['start']] = hg19_gtf['end']
    hg19_gtf.loc[hg19_gtf['strand'] == "-", ['end']] = hg19_gtf['start']+1
    hg19_gtf = hg19_gtf[["chr","start","end", "gene"]]

    hg19_gtf['gene'] = hg19_gtf['gene'].str.rstrip(';')


    return(hg19_gtf)

Coding_promoter_df = Identify_coding_promoters(gtf)
Coding_promoter = pybed.BedTool.from_dataframe(Coding_promoter_df)
Coding_promoter = Coding_promoter.sort()

MCF7_PcHiC = pd.read_table("/home/darragh/New/PcHiC/MCF7/round_2/Arima/chicago/data/MCF7.ibed", header=0)

#Focusing just on cis-cis interactions
MCF7_PcHiC = MCF7_PcHiC.loc[MCF7_PcHiC['bait_chr'] == MCF7_PcHiC['otherEnd_chr'] ]

MCF7_PcHiC = pybed.BedTool.from_dataframe(MCF7_PcHiC[['bait_chr',	'bait_start',	'bait_end', 'otherEnd_chr',	'otherEnd_start',	'otherEnd_end']])
MCF7_PcHiC_genes = MCF7_PcHiC.intersect(Coding_promoter, wa=True, wb=True)
MCF7_PcHiC_genes = pd.read_table(MCF7_PcHiC_genes.fn, header=None)
#This is the other end loop and its connected gene
MCF7_PcHiC_genes = MCF7_PcHiC_genes.iloc[:,[3,4,5,9]]
MCF7_PcHiC_genes.columns = ['chr', 'start', 'end', 'gene']
MCF7_PcHiC_genes = pybed.BedTool.from_dataframe(MCF7_PcHiC_genes)

MCF7_regs = pd.read_table("/home/darragh/New/Regulatory_elements/MCF7_reg_elements.txt", sep = "\t", header=None)
MCF7_regs.columns = ['chr', 'start', 'end', 'annotation']

MCF7_regs_AE = pybed.BedTool.from_dataframe(MCF7_regs.loc[MCF7_regs['annotation'] == "MCF7_active_enhancer"])
MCF7_regs_S = pybed.BedTool.from_dataframe(MCF7_regs.loc[MCF7_regs['annotation'] == "MCF7_H3K27me3_silencer"])

MCF7_regs_AE_genes = MCF7_regs_AE.intersect(MCF7_PcHiC_genes, wa = True, wb= True)
MCF7_regs_AE_genes = pd.read_table(MCF7_regs_AE_genes.fn, header=None)
MCF7_regs_AE_genes.columns = ['chr', 'start', 'end', 'anno', 'n1', 'n2','n3','gene']
MCF7_regs_AE_genes['id'] = MCF7_regs_AE_genes['chr'] + '_' + MCF7_regs_AE_genes['start'].astype(str) + '_' + MCF7_regs_AE_genes['end'].astype(str)

MCF7_regs_S_genes = MCF7_regs_S.intersect(MCF7_PcHiC_genes, wa = True, wb= True)
MCF7_regs_S_genes = pd.read_table(MCF7_regs_S_genes.fn, header=None)
MCF7_regs_S_genes.columns = ['chr', 'start', 'end', 'anno', 'n1', 'n2','n3','gene']
MCF7_regs_S_genes['id'] = MCF7_regs_S_genes['chr'] + '_' + MCF7_regs_S_genes['start'].astype(str) + '_' + MCF7_regs_S_genes['end'].astype(str)

MCF7_regs_AE_genes = MCF7_regs_AE_genes[['id', 'gene', 'anno']]
MCF7_regs_AE_genes = MCF7_regs_AE_genes.groupby(['anno', 'gene'])['id'].nunique().reset_index(name='count')

MCF7_regs_S_genes = MCF7_regs_S_genes[['id', 'gene', 'anno']]
MCF7_regs_S_genes = MCF7_regs_S_genes.groupby(['anno', 'gene'])['id'].nunique().reset_index(name='count')

MCF7_distal_regs = pd.concat([MCF7_regs_AE_genes, MCF7_regs_S_genes])

all_RNA = pd.read_table("/home/darragh/New/CCLE_data/OmicsExpressionProteinCodingGenesTPMLogp1.csv", sep = ",", header=0)
all_RNA = all_RNA.rename(columns={'Unnamed: 0': 'id'})

MCF7_RNA = all_RNA.loc[(all_RNA['id'] == 'ACH-000019') ]
MCF7_RNA = MCF7_RNA.T
MCF7_RNA.columns = MCF7_RNA.iloc[0]
MCF7_RNA = MCF7_RNA.iloc[1:]
MCF7_RNA = MCF7_RNA.reset_index()
MCF7_RNA.columns = ['gene', 'expression']
MCF7_RNA['gene'] = MCF7_RNA['gene'].str.split().str[0]
MCF7_RNA['z_score'] = (MCF7_RNA['expression'] - MCF7_RNA['expression'].mean() ) / MCF7_RNA['expression'].std()


MCF7_distal_regs_expression = MCF7_distal_regs.merge(MCF7_RNA, on="gene")
MCF7_distal_regs_expression = MCF7_distal_regs_expression.drop_duplicates()


#MCF7_distal_regs_expression = MCF7_distal_regs_expression.loc[MCF7_distal_regs_expression['expression'] > 0]

print(MCF7_distal_regs_expression.loc[MCF7_distal_regs_expression['anno'] == "MCF7_H3K27me3_silencer"]['expression'].median())
print(MCF7_distal_regs_expression.loc[MCF7_distal_regs_expression['anno'] == "MCF7_active_enhancer"]['expression'].median())

# Apply pd.qcut within each group, handling duplicates
MCF7_distal_regs_expression['quartile'] = MCF7_distal_regs_expression.groupby('anno')['count'].transform(
    lambda x: pd.qcut(x.rank(method='first'), q=3, labels=[1, 2, 3])
)

MCF7_distal_regs_expression['anno_quartile'] = MCF7_distal_regs_expression['anno'] + '_' + MCF7_distal_regs_expression['quartile'].astype(str)

display(MCF7_distal_regs_expression)

MCF7_distal_regs_expression.boxplot(column='z_score', by='anno_quartile', grid=False)

```

```python
####
#Function for calling promoters
####

import pandas as pd
import pybedtools as pybed

#awk '{print $0}' /index/hg19/Homo_sapiens.GRCh37.87.chr.gtf | grep "protein_coding" | awk '{if ($3 == "gene") print "chr"$1"\t"$4"\t"$5"\t"$7"\t"$14 }' > /home/darragh/Annotation_reg_elements/hg19_protein_coding_minimal.gtf
gtf = "/home/darragh/Annotation_reg_elements/hg19_minimal.gtf"
def Identify_coding_promoters(gtf):
    hg19_gtf = pd.read_table(gtf)
    hg19_gtf.columns = ["chr","start","end", "strand", "gene"]
    hg19_gtf = hg19_gtf[["chr","start","end", "gene"]]
    seqnames_we_like = ["chr1","chr2","chr3","chr4","chr5","chr6","chr7","chr8","chr9","chr10","chr11","chr12","chr13","chr14","chr15","chr16","chr17","chr18","chr19","chr20","chr21","chr22","chrX", "chrY"]
    hg19_gtf = hg19_gtf[hg19_gtf['chr'].isin(seqnames_we_like)]
    hg19_gtf['end'] = hg19_gtf['start']+1
    hg19_gtf['gene'] = hg19_gtf['gene'].str.rstrip(';')
    #hg19_gtf['start'] = hg19_gtf['start']-2500
    #hg19_gtf['end'] = hg19_gtf['end']+2500

    return(hg19_gtf)

Coding_promoter_df = Identify_coding_promoters(gtf)
display(Coding_promoter_df)
Coding_promoter = pybed.BedTool.from_dataframe(Coding_promoter_df)
Coding_promoter = Coding_promoter.sort()
```
