```r

####
#Function for calling promoters
####
#awk '{print $0}' /index/hg19/Homo_sapiens.GRCh37.87.chr.gtf | grep "protein_coding" | awk '{if ($3 == "gene") print "chr"$1"\t"$4"\t"$5"\t"$7"\t"$14 }' > /home/darragh/Annotation_reg_elements/hg19_protein_coding_minimal.gtf
gtf = "/home/darragh/Annotation_reg_elements/hg19_protein_coding_minimal.gtf"
def Identify_coding_promoters(gtf):
    hg19_gtf = pd.read_table(gtf)
    hg19_gtf.columns = ["chr","start","end", "strand", "gene"]
    hg19_gtf = hg19_gtf[["chr","start","end", "gene"]]
    seqnames_we_like = ["chr1","chr2","chr3","chr4","chr5","chr6","chr7","chr8","chr9","chr10","chr11","chr12","chr13","chr14","chr15","chr16","chr17","chr18","chr19","chr20","chr21","chr22","chrX", "chrY"]
    hg19_gtf = hg19_gtf[hg19_gtf['chr'].isin(seqnames_we_like)]
    hg19_gtf['end'] = hg19_gtf['start']+1
    hg19_gtf['gene'] = hg19_gtf['gene'].str.rstrip(';')

    return(hg19_gtf)

Coding_promoter_df = Identify_coding_promoters(gtf)
display(Coding_promoter_df)
Coding_promoter = pybed.BedTool.from_dataframe(Coding_promoter_df)
Coding_promoter = Coding_promoter.sort()
```

```r
HMEC = pybed.BedTool.from_dataframe(pd.read_table("/home/darragh/New/Regulatory_elements/HMEC_reg_elements.txt", sep="\t",header=None))
MCF10A = pybed.BedTool.from_dataframe(pd.read_table("/home/darragh/New/Regulatory_elements/MCF10A_reg_elements.txt", sep="\t",header=None))
MCF7 = pybed.BedTool.from_dataframe(pd.read_table("/home/darragh/New/Regulatory_elements/MCF7_reg_elements.txt", sep="\t",header=None))
T47D = pybed.BedTool.from_dataframe(pd.read_table("/home/darragh/New/Regulatory_elements/T47D_reg_elements.txt", sep="\t",header=None))
MDAMB231 = pybed.BedTool.from_dataframe(pd.read_table("/home/darragh/New/Regulatory_elements/MDAMB231_reg_elements.txt", sep="\t",header=None))
MDAMB436 = pybed.BedTool.from_dataframe(pd.read_table("/home/darragh/New/Regulatory_elements/MDAMB436_reg_elements.txt", sep="\t",header=None))
MDAMB468 = pybed.BedTool.from_dataframe(pd.read_table("/home/darragh/New/Regulatory_elements/MDAMB468_reg_elements.txt", sep="\t",header=None))
BT549 = pybed.BedTool.from_dataframe(pd.read_table("/home/darragh/New/Regulatory_elements/BT549_reg_elements.txt", sep="\t",header=None))

All = HMEC.cat(MCF10A, postmerge=False)
All = All.cat(MCF7, postmerge=False)
All = All.cat(T47D, postmerge=False)
All = All.cat(MDAMB231, postmerge=False)
All = All.cat(MDAMB436, postmerge=False)
All = All.cat(MDAMB468, postmerge=False)
All = All.cat(BT549, postmerge=False)

All = All.sort()
All = All.merge()

All = pd.read_table(All.fn, header=None)
All.columns = ['chr', 'start', 'end']
All['id'] = All['chr'] + '_' + All['start'].astype(str) + '_' + All['end'].astype(str)
All.to_csv("/home/darragh/New/ActiveDriverWGS/All_regs.txt", sep="\t", header=False, index=False)


display(All)
```

```r
#Test elements will be my complete set of active regulatory elements identified in all breast tissues (union)

MCF7 = pd.read_table("/home/darragh/New/Regulatory_elements/MCF7_reg_elements.txt", header=None, sep="\t")
T47D = pd.read_table("/home/darragh/New/Regulatory_elements/T47D_reg_elements.txt", header=None, sep="\t")

MCF7.columns = ['chr', 'start', 'end', 'annotation']
T47D.columns = ['chr', 'start', 'end', 'annotation']

ER_active_promoters = MCF7.loc[(MCF7['annotation'] == "MCF7_active_promoter")]
ER_active_promoters['annotation'] = "ER_active_promoter"

ER_polycomb_promoters = MCF7.loc[(MCF7['annotation'] == "MCF7_polycomb_promoter")]
ER_polycomb_promoters['annotation'] = "ER_polycomb_promoter"

MCF7_active_enhancers = pybed.BedTool.from_dataframe(MCF7.loc[(MCF7['annotation'] == "MCF7_active_enhancer")])

T47D_active_enhancers = pybed.BedTool.from_dataframe(T47D.loc[(T47D['annotation'] == "T47D_active_enhancer")])

ER_active_enhancers = MCF7_active_enhancers.cat(T47D_active_enhancers, postmerge=False)

ER_active_enhancers = ER_active_enhancers.sort()

ER_active_enhancers = ER_active_enhancers.merge()

ER_active_enhancers = pd.read_table(ER_active_enhancers.fn, header=None)

ER_active_enhancers.columns = ['chr', 'start', 'end']
ER_active_enhancers['annotation'] = "ER_active_enhancer"



ER_regs = pd.concat([ER_active_promoters, ER_polycomb_promoters, ER_active_enhancers])
ER_regs['id'] = ER_regs['chr'] + '_' + ER_regs['start'].astype(str) + '_' + ER_regs['end'].astype(str)
ER_regs = ER_regs.drop_duplicates()
ER_regs = ER_regs[['chr', 'start', 'end', 'id']]
ER_regs.to_csv("/home/darragh/New/ActiveDriverWGS/inputs/ER_regs.txt", sep="\t", index=False, header=None)
display(ER_regs)
```
