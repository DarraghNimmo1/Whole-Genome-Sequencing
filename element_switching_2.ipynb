```python
import pandas as pd
import pybedtools as pybed
import glob
import os
import numpy as np

#MCF7 ATAC
MCF7_rep1_ATAC_enh_to_sil_coverage = pd.read_table("/home/darragh/New/Differentially_active/switching/MCF7/coverage/MCF7_ATAC_rep1_sil_to_enh_coverage_2.txt", sep = "\t")
MCF7_rep1_ATAC_enh_to_sil_coverage.columns = ['chr', 'start', 'end', 'annotation', 'count_1']
MCF7_rep1_ATAC_enh_to_sil_coverage['count_1'] = (MCF7_rep1_ATAC_enh_to_sil_coverage['count_1']/51944564)*1000000
MCF7_rep2_ATAC_enh_to_sil_coverage = pd.read_table("/home/darragh/New/Differentially_active/switching/MCF7/coverage/MCF7_ATAC_rep2_sil_to_enh_coverage_2.txt", sep = "\t")
MCF7_rep2_ATAC_enh_to_sil_coverage.columns = ['chr', 'start', 'end', 'annotation', 'count_2']
MCF7_rep2_ATAC_enh_to_sil_coverage['count_2'] = (MCF7_rep2_ATAC_enh_to_sil_coverage['count_2']/45876682)*1000000
MCF7_ATAC_enh_to_sil_coverage = MCF7_rep1_ATAC_enh_to_sil_coverage.merge(MCF7_rep2_ATAC_enh_to_sil_coverage)
MCF7_ATAC_enh_to_sil_coverage['ATAC_MCF7'] = (MCF7_ATAC_enh_to_sil_coverage['count_1'] + MCF7_ATAC_enh_to_sil_coverage['count_2'])/2
MCF7_ATAC_enh_to_sil_coverage = MCF7_ATAC_enh_to_sil_coverage[['chr',	'start',	'end',	'annotation', 'ATAC_MCF7']]

#MCF7 H3K27ac
MCF7_H3K27ac_enh_to_sil_coverage = pd.read_table("/home/darragh/New/Differentially_active/switching/MCF7/coverage/MCF7_H3K27ac_sil_to_enh_coverage_2.txt", sep = "\t")
MCF7_H3K27ac_enh_to_sil_coverage.columns = ['chr', 'start', 'end', 'annotation', 'H3K27ac_MCF7']
MCF7_H3K27ac_enh_to_sil_coverage['H3K27ac_MCF7'] = (MCF7_H3K27ac_enh_to_sil_coverage['H3K27ac_MCF7']/24282555)*1000000
MCF7_enh_to_sil_coverage = MCF7_ATAC_enh_to_sil_coverage.merge(MCF7_H3K27ac_enh_to_sil_coverage)

#MCF7 H3K27me3
MCF7_H3K27me3_enh_to_sil_coverage = pd.read_table("/home/darragh/New/Differentially_active/switching/MCF7/coverage/MCF7_H3K27me3_sil_to_enh_coverage_2.txt", sep = "\t")
MCF7_H3K27me3_enh_to_sil_coverage.columns = ['chr', 'start', 'end', 'annotation', 'H3K27me3_MCF7']
MCF7_H3K27me3_enh_to_sil_coverage['H3K27me3_MCF7'] = (MCF7_H3K27me3_enh_to_sil_coverage['H3K27me3_MCF7'] / 24590803) *1000000 
MCF7_enh_to_sil_coverage = MCF7_enh_to_sil_coverage.merge(MCF7_H3K27me3_enh_to_sil_coverage)

#MCF7 H3K4me1
MCF7_H3K4me1_enh_to_sil_coverage = pd.read_table("/home/darragh/New/Differentially_active/switching/MCF7/coverage/MCF7_H3K4me1_sil_to_enh_coverage_2.txt", sep = "\t")
MCF7_H3K4me1_enh_to_sil_coverage.columns = ['chr', 'start', 'end', 'annotation', 'H3K4me1_MCF7']
MCF7_H3K4me1_enh_to_sil_coverage['H3K4me1_MCF7'] = (MCF7_H3K4me1_enh_to_sil_coverage['H3K4me1_MCF7'] / 24720826) *1000000 
MCF7_enh_to_sil_coverage = MCF7_enh_to_sil_coverage.merge(MCF7_H3K4me1_enh_to_sil_coverage)

#MCF7 SUZ12
MCF7_SUZ12_enh_to_sil_coverage = pd.read_table("/home/darragh/New/Differentially_active/switching/MCF7/coverage/MCF7_SUZ12_sil_to_enh_coverage_2.txt", sep = "\t")
MCF7_SUZ12_enh_to_sil_coverage.columns = ['chr', 'start', 'end', 'annotation', 'SUZ12_MCF7']
MCF7_SUZ12_enh_to_sil_coverage['SUZ12_MCF7'] = (MCF7_SUZ12_enh_to_sil_coverage['SUZ12_MCF7'] / 20941228) *1000000 
MCF7_enh_to_sil_coverage = MCF7_enh_to_sil_coverage.merge(MCF7_SUZ12_enh_to_sil_coverage)

#HMEC ATAC
HMEC_rep1_ATAC_enh_to_sil_coverage = pd.read_table("/home/darragh/New/Differentially_active/switching/MCF7/coverage/HMEC_rep1_ATAC_sil_to_enh_coverage_2.txt", sep = "\t")
HMEC_rep1_ATAC_enh_to_sil_coverage.columns = ['chr', 'start', 'end', 'annotation', 'count_1']
HMEC_rep1_ATAC_enh_to_sil_coverage['count_1'] = (HMEC_rep1_ATAC_enh_to_sil_coverage['count_1']/125012258)*1000000
HMEC_rep2_ATAC_enh_to_sil_coverage = pd.read_table("/home/darragh/New/Differentially_active/switching/MCF7/coverage/HMEC_rep2_ATAC_sil_to_enh_coverage_2.txt", sep = "\t")
HMEC_rep2_ATAC_enh_to_sil_coverage.columns = ['chr', 'start', 'end', 'annotation', 'count_2']
HMEC_rep2_ATAC_enh_to_sil_coverage['count_2'] = (HMEC_rep2_ATAC_enh_to_sil_coverage['count_2']/144302678)*1000000
HMEC_ATAC_enh_to_sil_coverage = HMEC_rep1_ATAC_enh_to_sil_coverage.merge(HMEC_rep2_ATAC_enh_to_sil_coverage)
HMEC_ATAC_enh_to_sil_coverage['ATAC_HMEC'] = (HMEC_ATAC_enh_to_sil_coverage['count_1'] + HMEC_ATAC_enh_to_sil_coverage['count_2'])/2
HMEC_ATAC_enh_to_sil_coverage = HMEC_ATAC_enh_to_sil_coverage[['chr',	'start',	'end',	'annotation', 'ATAC_HMEC']]

#HMEC H3K27ac
HMEC_H3K27ac_enh_to_sil_coverage = pd.read_table("/home/darragh/New/Differentially_active/switching/MCF7/coverage/HMEC_H3K27ac_sil_to_enh_coverage_2.txt", sep = "\t")
HMEC_H3K27ac_enh_to_sil_coverage.columns = ['chr', 'start', 'end', 'annotation', 'H3K27ac_HMEC']
HMEC_H3K27ac_enh_to_sil_coverage['H3K27ac_HMEC'] = (HMEC_H3K27ac_enh_to_sil_coverage['H3K27ac_HMEC']/21528652)*1000000
HMEC_enh_to_sil_coverage = HMEC_ATAC_enh_to_sil_coverage.merge(HMEC_H3K27ac_enh_to_sil_coverage)

#HMEC H3K27me3
HMEC_H3K27me3_enh_to_sil_coverage = pd.read_table("/home/darragh/New/Differentially_active/switching/MCF7/coverage/HMEC_H3K27me3_sil_to_enh_coverage_2.txt", sep = "\t")
HMEC_H3K27me3_enh_to_sil_coverage.columns = ['chr', 'start', 'end', 'annotation', 'H3K27me3_HMEC']
HMEC_H3K27me3_enh_to_sil_coverage['H3K27me3_HMEC'] = (HMEC_H3K27me3_enh_to_sil_coverage['H3K27me3_HMEC']/21063005)*1000000
HMEC_enh_to_sil_coverage = HMEC_enh_to_sil_coverage.merge(HMEC_H3K27me3_enh_to_sil_coverage)
enh_to_sil_coverage = HMEC_enh_to_sil_coverage.merge(MCF7_enh_to_sil_coverage)

#HMEC H3K4me1
HMEC_H3K4me1_enh_to_sil_coverage = pd.read_table("/home/darragh/New/Differentially_active/switching/MCF7/coverage/HMEC_H3K4me1_sil_to_enh_coverage_2.txt", sep = "\t")
HMEC_H3K4me1_enh_to_sil_coverage.columns = ['chr', 'start', 'end', 'annotation', 'H3K4me1_HMEC']
HMEC_H3K4me1_enh_to_sil_coverage['H3K4me1_HMEC'] = (HMEC_H3K4me1_enh_to_sil_coverage['H3K4me1_HMEC']/29228426)*1000000
HMEC_enh_to_sil_coverage = HMEC_enh_to_sil_coverage.merge(HMEC_H3K4me1_enh_to_sil_coverage)
enh_to_sil_coverage = HMEC_enh_to_sil_coverage.merge(MCF7_enh_to_sil_coverage)

#HMEC SUZ12
HMEC_SUZ12_enh_to_sil_coverage = pd.read_table("/home/darragh/New/Differentially_active/switching/MCF7/coverage/HMEC_SUZ12_sil_to_enh_coverage_2.txt", sep = "\t")
HMEC_SUZ12_enh_to_sil_coverage.columns = ['chr', 'start', 'end', 'annotation', 'SUZ12_HMEC']
HMEC_SUZ12_enh_to_sil_coverage['SUZ12_HMEC'] = (HMEC_SUZ12_enh_to_sil_coverage['SUZ12_HMEC']/16180774)*1000000
HMEC_enh_to_sil_coverage = HMEC_enh_to_sil_coverage.merge(HMEC_SUZ12_enh_to_sil_coverage)
enh_to_sil_coverage = HMEC_enh_to_sil_coverage.merge(MCF7_enh_to_sil_coverage)


enh_to_sil_coverage['ATAC_FC'] = np.log2((enh_to_sil_coverage['ATAC_MCF7']+1) / (enh_to_sil_coverage['ATAC_HMEC']+1))
enh_to_sil_coverage['H3K27ac_FC'] = np.log2((enh_to_sil_coverage['H3K27ac_MCF7']+1) / (enh_to_sil_coverage['H3K27ac_HMEC']+1))
enh_to_sil_coverage['H3K27me3_FC'] = np.log2((enh_to_sil_coverage['H3K27me3_MCF7']+1) / (enh_to_sil_coverage['H3K27me3_HMEC']+1))
enh_to_sil_coverage['H3K4me1_FC'] = np.log2((enh_to_sil_coverage['H3K4me1_MCF7']+1) / (enh_to_sil_coverage['H3K4me1_HMEC']+1))
enh_to_sil_coverage['SUZ12_FC'] = np.log2((enh_to_sil_coverage['SUZ12_MCF7']+1) / (enh_to_sil_coverage['SUZ12_HMEC']+1))

enh_to_sil_coverage = enh_to_sil_coverage[['chr',	'start',	'end',	'annotation', 'ATAC_FC', 'H3K27ac_FC', 'H3K27me3_FC', 'H3K4me1_FC', 'SUZ12_FC'  ]]
enh_to_sil_coverage = enh_to_sil_coverage.drop_duplicates()
enh_to_sil_coverage.to_csv("/home/darragh/New/Differentially_active/switching/MCF7/output_sil_to_enh_2.txt", sep = "\t", index=False)
display(enh_to_sil_coverage)

```
